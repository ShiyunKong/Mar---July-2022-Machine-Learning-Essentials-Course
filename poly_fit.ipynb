```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/poly_fit.csv")
print(df)
```

```python
x = df.iloc[:,0]
y = df.iloc[:,1]
```

```python
import numpy as np

```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=5, include_bias=False)
from sklearn import linear_model
x_reshape =np.array(x).reshape((-1,1))
poly_feature = poly.fit_transform(x_reshape)
poly_reg_model = linear_model.LinearRegression()
poly_reg_model.fit(poly_feature, y)
y_predicted = poly_reg_model.predict(poly_feature)

# creating the plot
x0 = np.linspace(0, 18 , 100).reshape((-1,1))
y0 = poly_reg_model.predict(poly.transform(x0))
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(x0,y0, label='fitted line', c='red')
plt.scatter(x, y, label='true points', c='blue')
plt.legend() #showing the legend
plt.show()
```

```python
import pandas as pd
daf = pd.DataFrame(y0)
daf.to_csv("/content/drive/MyDrive/Colab Notebooks/poly_fit_regress.csv", index=False)
```
